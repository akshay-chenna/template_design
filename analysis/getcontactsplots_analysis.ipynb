# Origin domain contacts fractions

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = pd.read_table('nsp12_contactresfreq.txt', sep='-', header=None)
y = pd.read_table('p4_contactresfreq.txt', sep='-', header=None)
z = pd.read_table('p5_1_contactresfreq.txt', sep='-', header=None)
```

```python
frac_x = x[1]/8334
frac_y = y[1]/1e4
frac_z = z[1]/1e4
```

```python
nsp12_p4 = np.vstack((frac_x[:8],frac_y)).T
```

```python
nsp12_p5 = np.vstack((frac_x[2:],frac_z)).T
```

```python
sns.heatmap(nsp12_p4,cmap='bwr', xticklabels=['nsp12','P4'], yticklabels=list(y[0]))
plt.text(2.05,5.5,'Contact frequency', rotation=90)
plt.savefig('nsp12_P4_contact.jpg', dpi=600, bbox_tight='True')
```

```python
sns.heatmap(nsp12_p5,cmap='bwr', xticklabels=['nsp12','P5'], yticklabels=list(z[0]))
plt.text(2.05,7,'Contact frequency', rotation=90)
#plt.savefig('nsp12_P5_contact.jpg', dpi=600, bbox_tight='True')
```
