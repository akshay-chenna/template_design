## Compute the total binding energies, ie include the interaction entropy term

## $\epsilon$ = 4

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

files = open('mmlist.txt')
lines = files.readlines()

be_all = np.empty(len(lines))
xmmpbsa = np.empty(len(lines))
mmpbsa_noie = np.empty(len(lines))
count = 0
for line in lines:
    
    mmpbsa = np.genfromtxt(str(line[:-1])+'/mmpbsa4.txt') # Index, vdw, elec, delta_pol, delta_apol, sum

    R = 8.314
    T = 310
    betainv = R*T/1000

    mmpbsa[:,2] = mmpbsa[:,2]
    mm = mmpbsa[:,1]+mmpbsa[:,2]
    ie = betainv*np.log(np.mean(np.exp((mm-np.mean(mm))/betainv)))
    be = np.mean(mmpbsa[:,-1])+ie
    xmmpbsa[count] = np.mean(mmpbsa[:,1])/4.2
    mmpbsa_noie[count] = np.mean(mmpbsa[:,-1])/4.2
    be_all[count] = be/4.2
    count += 1
```

```python
np.save('mmpbsa4_noie.npy',mmpbsa_noie.reshape(20,5))
```
