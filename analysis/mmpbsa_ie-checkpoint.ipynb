# Compute the total binding energies, ie include the interaction entropy term

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
files = open('mmlist.txt')
lines = files.readlines()

be_all = np.empty(len(lines))
count = 0
for line in lines:
    
    mmpbsa = np.genfromtxt(str(line[:-1])+'/mmpbsa1.txt')

    R = 8.314
    T = 310
    betainv = R*T/1000

    mmpbsa[:,2] = mmpbsa[:,2]
    mm = mmpbsa[:,1]+mmpbsa[:,2]
    ie = betainv*np.log(np.mean(np.exp((mm-np.mean(mm))/betainv)))
    be = np.mean(mmpbsa[:,-1])+ie
    be_all[count] = be/4.2
    count += 1
fasc = np.genfromtxt('reu.txt')[:,0]
```

```python
abe_all1 = np.load('abe1.npy')
afasc1 = np.load('areu.npy')


be_all = np.concatenate((be_all,abe_all1))
fasc = np.concatenate((fasc,afasc1))
```

```python
c = np.concatenate((np.zeros(1),np.ones(87),np.zeros(1)))
#t = 2*np.ones(58)
#t[0] = 3
#t[16] = 3
#t[31] = 3

plt.figure(figsize=(6,4)) 
plt.rcParams.update({'font.size': 10})
sb.scatterplot(fasc,be_all, hue=c, s=100, alpha=0.6, edgecolor='k', palette='Set1', legend=1)
plt.legend(['_', 'L','L+D'])
plt.xlabel('Design: REU (kcal/mol)')
plt.ylabel('MD: MMPBSA+IE (kcal/mol)')
plt.title('Comparing binding energies of probable CPP designs and MD $\epsilon$ = 1')
plt.text(fasc[88]+0.2,be_all[88]+5,'native')
x = np.arange(-45,-5,0.01)
plt.plot(x,x,'g')
plt.savefig('be-reu_dielectric1.jpg', dpi=600)
```

```python
np.save('be1',be_all)
np.save('fasc',fasc)
```

## $\epsilon$ = 4

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

files = open('mmlist.txt')
lines = files.readlines()

be_all = np.empty(len(lines))
xmmpbsa = np.empty(len(lines))
count = 0
for line in lines:
    
    mmpbsa = np.genfromtxt(str(line[:-1])+'/mmpbsa4.txt')

    R = 8.314
    T = 310
    betainv = R*T/1000

    mmpbsa[:,2] = mmpbsa[:,2]
    mm = mmpbsa[:,1]+mmpbsa[:,2]
    ie = betainv*np.log(np.mean(np.exp((mm-np.mean(mm))/betainv)))
    be = np.mean(mmpbsa[:,-1])+ie
    xmmpbsa[count] = np.mean(mmpbsa[:,1])/4.2
    be_all[count] = be/4.2
    count += 1
fasc = np.genfromtxt('reu.txt')[:,0]

abe_all1 = np.load('abe4.npy')
afasc1 = np.load('areu.npy')

np.savetxt('2be_mmpbsa.txt', xmmpbsa, fmt='%.2f')

be_all = np.concatenate((be_all,abe_all1))
fasc = np.concatenate((fasc,afasc1))

c = np.concatenate((np.zeros(1),np.ones(87),np.zeros(1)))

plt.figure(figsize=(6,4)) 
plt.rcParams.update({'font.size': 14})
sb.scatterplot(fasc,be_all, hue=c, s=100, alpha=0.6, edgecolor='k', palette='Set1', legend=1)
plt.legend(['_', 'L','L+D'])
plt.xlabel('Design: REU (kcal/mol)')
plt.ylabel('MD: MMPBSA+IE (kcal/mol)')

#plt.xlabel('REU (kcal/mol)')
#plt.ylabel('MMPBSA+IE (kcal/mol)')
#plt.title('Comparing binding energies of probable CPP designs and MD $\epsilon$ = 4')
plt.text(fasc[88]+0.1,be_all[88]+2,'native')
x = np.arange(-45,20,0.01)
plt.plot(x,x,'g')
#plt.savefig('be-reu_dielectric4.jpg', dpi=600)
plt.xlim([-45,22])
plt.ylim([-45,22])

#np.save('be4',be_all)
#np.savetxt('be4_reu_cpp.txt',(be_all,fasc), '%.3f')
be4 = be_all
plt.savefig('be-reu_dielectric4.jpg', dpi=600)
```

```python
import numpy as np
be4 = np.load('be4.npy')
reu = np.load('fasc.npy')
best_mmpbsa = np.where(be4 < -30)
best_reu = np.where(reu < -10)
selected = np.intersect1d(best_mmpbsa, best_reu) + 1
np.savetxt('topbe-reu_poses.txt',selected, '%d')
```
